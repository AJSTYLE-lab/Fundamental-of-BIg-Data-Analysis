```python
# K mean - case 1
# Code which output time, purity, distiniugable plot 
#conclusive line
X, y = make_moons(n_samples=7692, random_state=42,noise=0.05)
LIST = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    LIST.append(kmeans.inertia_)
plt.plot(range(1, 10), LIST, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters----->')
plt.ylabel('Inertia/Frequency------>')
plt.show()
start_time = time.time()
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
end_time = time.time()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
cluster1 = np.where(y== 0)[0]
cluster2 = np.where(y== 1)[0]
purity = max(np.sum(y[cluster1] == 0) + np.sum(y[cluster2] == 1), np.sum(y[cluster1] == 1) + np.sum(y[cluster2] == 0)) / len(y)
print('Time taken:', end_time - start_time, 'seconds')
print('Purity:', purity)
```

```python
# Dbscan - case 1
# Code which output time, purity, distinguishable plot
#conclusive line
from sklearn.cluster import DBSCAN
X, y = make_moons(n_samples=6298, noise=0.05)
X = StandardScaler().fit_transform(X)
START = time.time()
db = DBSCAN(eps=0.3, min_samples=20).fit(X)
y_pred_label=db.fit_predict(X)
masks = db.labels_ 
END = time.time()
plt.scatter(X[:, 0], X[:, 1], c=masks)
plt.title("DBSCAN Clustering")
plt.show()
labels_true = y
print('Time taken:', END- START, 'seconds')
labels_true = y
labels_pred =y_pred_label
equal = {}
unique_labels = set(labels_pred)
for masks in unique_labels:
    equal[masks] = np.argmax(np.bincount(labels_true[labels_pred == masks]))
y_predicted = [equal[masks] for masks in y_pred_label]
purity = accuracy_score(y, y_predicted)
print("purity=",purity)
```

#CONCLUSIVE LINE FOR DB_SCAN  case-1:
THIS ALGO is  very fast on  dataset as compare to others because it takes 0.17 seconds with highest purity of 1.00, which is max purity/entropy.
