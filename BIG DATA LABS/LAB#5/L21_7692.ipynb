1->IMPORTING LIBRARIES:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN 
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

<a id = "2"></a>
2->MAKING POPULATION/DATA POINTS:


```python

X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
print(X,y)
```

<a id = "3"></a>
3->DB_SCAN:

```python
dbscan = DBSCAN(eps=0.5, min_samples=5) # create an instance of DBSCAN with parameters
epsilon=0.5
min_samples=5
dbscan.fit(X) # fit DBSCAN model to the data
# Get cluster labels
labels = dbscan.labels_ # get cluster labels from DBSCAN model
```

<a id = "4"></a>
4->PLOTTING RESULT:

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow') # plot the data points with different colors for different clusters
plt.title('DBSCAN Clustering') # set the title of the plot
plt.show() # show the plot
```

GENERATIND DATA POINTS/POPULATION:


```python
X, Y = make_blobs(n_samples=1000, centers=3, n_features=2,random_state=0)
X = StandardScaler().fit_transform(X)
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
print(X,Y)
```

ARRAY OF 0 WITH SAME SHAPE+MASKING/LABELING:

```python
SAMPLE_MASKING = np.zeros_like(db.labels_, dtype=bool)
SAMPLE_MASKING[db.core_sample_indices_] = True
labels = db.labels_
labels
```

DOING CLUSTER LABELING FOR ALL DATA POINTS GIVEN TO FIT().OUTLIERS LABEL AS -1.

```python
unique_labels = set(labels)
unique_labels
```

DB_SCAN ALGORITHM:

```python
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]
    class_member_mask = (labels == k)
    xy = X[class_member_mask & SAMPLE_MASKING]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor='pink',
             markeredgecolor='k', markersize=14)
    xy = X[class_member_mask & ~SAMPLE_MASKING]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor='red',
             markeredgecolor='w', markersize=6)
plt.title("DB_SCAN CLUSTERING.")
plt.show()
```
