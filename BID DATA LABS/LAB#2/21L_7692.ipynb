#TASK C:

```python
X = np.array(number_file['Age'])
y = np.array(number_file['height'])
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.4, random_state=1)
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
# making predictions on the testing set
y_pred = knn.predict(X_test.reshape(-1, 1))
# comparing actual response values (y_test) with predicted
from sklearn import metrics
print("kNN model accuracy:", metrics.accuracy_score(y_test,
y_pred))
```
